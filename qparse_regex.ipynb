# Utility functions

```python
# Substitute unicode characters found in Citizen Quotes corpus
def clean_unicode_lines(text):
    line = text.encode('utf-8').strip()
    line = re.sub(r'\xc2\xa0', ' ', line)            # NO-BREAK SPACE
    line = re.sub(r'\xe2\x80(\x94|\x93)', '-', line) # EM DASH
    line = re.sub(r'\xe2\x80\xa6', '...', line)      # HORIZONTAL ELLIPSIS
    line = re.sub(r'\xe2\x80\xb3', '"', line)        # DOUBLE PRIME
    return line

quote_regex = re.compile(r'(\".+?\")')
```

# Function to separate quotes and non-quotes at the paragraph level

```python
def quote_print(document, prob_idx, text_idx):
    for key, para in enumerate(document):

        # probability paragraph containing a quote
        if para[prob_idx] > .5:
            paragraph = clean_unicode_lines(para[text_idx])

            # return list of text between quotation marks and indices
            quotes = quote_regex.findall(paragraph)
            quotes_idx = [q.start() for q in quote_regex.finditer(paragraph)]

            # return non-empty, non-quotes and indices
            delim_pattern = '|'.join(map(re.escape, quotes))
            get_quote_outside = re.split(delim_pattern, paragraph)
            nonquotes = filter(len, get_quote_outside)
            nonquotes_idx = [paragraph.index(nq) for nq in nonquotes]
            
            # combine indices for both quotes and non-quotes
            data = zip(nonquotes_idx, ['NON-QUOTE'] * len(nonquotes), nonquotes) + \
                   zip(quotes_idx, ['QUOTE'] * len(quotes), quotes)

            # sort quotes and nonquotes by index
            sorted_data = sorted(data, key=lambda x: x[0])
            for x in sorted_data:
                print '{:<8}{:<12}{}'.format(key, x[1], x[2])
            print

        # paragraphs with no quotes
        else:
            print '{:<8}{:<12}\n'.format(key, 'NONE')
```

```python
#chunker rule to rip the quote out
#just the quote
plainqt = 'QT: { (<``>|<\'\'>)+  <.*>+ (<''>|<\'\'>) }'
#everything in front of the quote plus the quote
qtbegin = 'QT: { <.*>+ (<``>|<\'\'>)+  <.*>+ (<''>|<\'\'>) }'
#everything after the quote
qtend = 'QT: { (<``>|<\'\'>)+  <.*>+ (<''>|<\'\'>)+ <.*>+ }'

#now try chinking- aka exclude patterns
#grab everything before or after the quote and then discard the quote
qtbeginstuff = '''QT: 
                    { <.*>+ (<``>|<\'\'>)+  <.*>+ (<''>|<\'\'>) }
                    }(<``>|<\'\'>)+ <.*>+ (<''>|<\'\'>){                    
                '''
qtendstuff = '''QT: 
                   { (<``>|<\'\'>)+  <.*>+ (<''>|<\'\'>)+ <.*>+ }
                   }(<``>|<\'\'>)+ <.*>+ (<''>|<\'\'>){  
             '''

#try to parse common patterns after a quote
#find the peep after a quote: ex: said Dick Chaney
qtendpeepsaidafter = '''QT: 
                       { (<``>|<\'\'>)+  <.*>+ (<''>|<\'\'>)+ <.*>+ }
                       }(<``>|<\'\'>)+ <.*>+ (<''>|<\'\'>){ 
                       {(<states>|<stated>|<said>|<expressed>|<told>|<exclaimed>|<announced>|<added>|<comment>)+}
                 '''

#get the easiest stuff first out of the start and end of the quotes...need to refine this more
peepsbeginqt = '''QT: 
                    { (<DT>?<RB>?)?<JJ|CD>*(<JJ|CD><,>)*(<NN.*>)+ <VB.*> <,>?  } 
                    } <VB.*> {
                    } <,> {
                '''

#get all the possible noun phrases after the end quote. 
peepsendqt = '''QT: 
                    { (<DT>?<RB>?)?<JJ|CD>*(<JJ|CD><,>)*(<NN.*>)+ <,>?  } 
                '''


#chunk and parse, and then put the remaining words back to together
def chunktext(sentencelist, rule):
    peepsdict = []
    for item in sentencelist:
      for sent in item[1]:
            itemdict = []
            cp = nltk.RegexpParser(rule)
            tree = cp.parse(sent)
            for subtree in tree.subtrees():
                if subtree.label() == 'QT': 
                 words =  [word[0] for word in subtree]
                 words = " ".join(words)
                 itemdict.append(words)
            if len(itemdict) > 0:
                peep = (item[0], " ".join(itemdict)) 
                peepsdict.append(peep)
    peepsdict = list(set(peepsdict))
    return peepsdict
```
