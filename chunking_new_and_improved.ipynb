```python
#############PARSING FUNCTIONS##################


#function to chunk text based on a rule
def chunktext(txt, rule):
    cp = nltk.RegexpParser(rule)
    tree = cp.parse(txt)
    peepsdict = []
    for subtree in tree.subtrees():
        itemdict = []
        if subtree.label() == 'QT': 
            tags  =  [word[1] for word in subtree]
            if 'NNP' in tags or 'COREF' in tags:
                words =  [word[0] for word in subtree]
                exclude = ['during', 'that','outside', 'now', 'this', 'down', 'because', 'about', 'with', 'about', 'each', 'later', 'during', 'just' ]
                cool =  set(words).isdisjoint(exclude)
                if cool:
                    words = " ".join(words)
                    itemdict.append(words)
    stuff = set(itemdict)
    for s in stuff:
        s = unidecode(s)
        s = str(s).strip().replace(".", "")
        if s not in peepsdict:
            peepsdict.append(s)
    return peepsdict


#cleans up weird garbage that are too hard to get out any way.
def clean_up(ner_list):
    clean = True
    words_to_exclude = ['for', 'in', 'of', 'by', 'outside', 'into', 'now', 'to', 'at', 'of' ',', 'on', 'this', 'down', 'because', 'about'
                        'with', 'about', 'each', 'later', 'during', 'just']
    stuff_list = ner_list.split(" ")
    if stuff_list[0] in  words_to_exclude:
        clean =  False
    else:
        stuff_list = " ".join(stuff_list)
        stuff_list = stuff_list.split(",")
        if len(stuff_list) > 2 :
            item1 = stuff_list[0].rstrip().lstrip()
            item2 = stuff_list[1].rstrip().lstrip()
            st_fuzz = fuzz.ratio(item1, item2) 
            if st_fuzz < 60:
                clean = False
    return clean


def get_final_results(df_item, rule, people, story, para):
    items = []
    for row in df_item.T.iteritems():
        to_chunk = row[1][6]
        peep = chunktext(to_chunk, rule )
        
        if not (peep is None):
            pz =  " ".join(peep)
            pz = pz.replace(".", "").rstrip().lstrip()
            if pz != '':
                rrow = []
                if clean_up(pz):
                    people.append(pz)
                    para.append(row[1][1])
                    story.append(row[1][2])
                    items.append(rrow)
    #df_said_done = pd.DataFrame({'story_id': np.array(finalresults),'para_index':np.array(para_indz), 'speaker': np.array(finalresults)}, columns=['story_id','para_index', 'speaker'])
    return people, para, story
                
def merge_results_to_main(df_said_done, df_all_done):
    #df_all_done = pd.concat(df_said_done, df_all_done)
    df_all_done = df_all_done.append(df_said_done,ignore_index = True)
    return df_all_done
```

```python
#function to generate the final output json files
def get_title_orgs( attrition_word, final_df ,rule,people, para, story):
    
    #flag all the paragraphs that contain said and then dump them into into main dataframe
    bqs_sts_qts[attrition_word+'_sent'] = pd.DataFrame(bqs_sts_qts.full_sent.str.contains(attrition_word+'+'))
    df_said =  bqs_sts_qts[ bqs_sts_qts[attrition_word+'_sent'] == True]
    df_said_done = get_final_results(df_said, rule, people, para, story)
    #print "*****Reduction/Hit/Parse Info****"
    #print "original df_qt size: " + str(len(bqs_sts_qts))
    #print "df attrition size: " +  str(len(df_said))
   # print "df title_org_parse size: " + str(len(df_said_done))
   # print "****************"
    #df_all_out = merge_results_to_main(df_said_done, df_all_done)
    return df_said_done
```
