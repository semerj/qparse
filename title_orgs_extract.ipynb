```python
potential_person =  '''
         PEEP: {<DT>?(<NNP*>|<COREF>)+<JJ.*>* <JJ.*>*<NN.*|CD>*}
'''
```

```python
def chunktext(txt, rule, person):
    cp = nltk.RegexpParser(rule)
    tree = cp.parse(txt)
    peepsdict = []
    for subtree in tree.subtrees():
        itemdict = []
        word_len = 0
        if subtree.label() == 'PEEP': 
            tags  =  [word[1] for word in subtree]
            words =  [word[0] for word in subtree]
            if 'that' not in words:
                words =  " ".join(words)
                fuzz_ratio = fuzz.partial_ratio(person, word)
                if fuzz_ratio > 50:
                    peepsdict.append(words)
    peepsdict = list(set(peepsdict))
    return peepsdict
```

```python
def get_story_text(story_id):
    spk2 = bqs[bqs['story_id'] == story_id]
    spk_sents = spk2['eval_tagged']
    text =spk_sents.as_matrix()
    article_text = list(itertools.chain(*text))
    return article_text
```

```python
text = ''
para_index = []
paragraph = []
quotations = []
quote_in_para = []
speaker = []
story_id = []
speaker_name = []
speaker_title_org = []
for c in cool:
    pi = c[0]
    pg = c[1]
    qt = c[2]
    qtp = c[3]
    sp = c[4]
    sty_id = c[5]
    title_org = ''
    sp_name = ''
    if not (sp is None) and sp[0] != 'Robert J. Glushko':
        person = sp[0]
        text =  get_story_text(sty_id)
        peep = chunktext(text,potential_person, person)
        if len(peep) > 0:
            np_peep_ct = max(len(p) for p in peep)
            np_peep = [p for p in peep if len(p) == np_peep_ct]
            speaker_stuff = np_peep[0].split("__")
            if len( speaker_stuff) > 1:
                title_org = speaker_stuff[0]
                sp_name = speaker_stuff[1]
            else:
                title_org = speaker_stuff[0]
    para_index.append(pi)
    paragraph.append(pg)
    quotations.append(qt)
    quote_in_para.append(qtp)
    speaker.append(sp)
    story_id.append(sty_id)
    speaker_name.append(sp_name)
    speaker_title_org.append(title_org)
```

```python
new_data = {'para_index': para_index,'paragraph': paragraph, 'quotations':quotations, 'quote_in_para': quote_in_para, 'speaker':speaker, 'story_id':story_id,  
            'speaker_name' : speaker_name, 'speaker_title_org':  speaker_title_org}
columns_sp = ['para_index','paragraph', 'quotations', 'quote_in_para', 'speaker', 'story_id',  
            'speaker_name', 'speaker_title_org']
```
