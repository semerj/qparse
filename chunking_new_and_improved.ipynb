```python
def find_person(text):
    words = [ word[0] for word in text]
    if 'said' in words:
        return True
    

#function to chunk text based on a rule
def chunktext(text, rule):
    peepsdict = []
   # words = pos_tag(word_tokenize(text))
    itemdict = []
    cp = nltk.RegexpParser(rule)
    for sent in text:
        sent = pos_tag(word_tokenize(sent))
        saidqt = find_person(sent)
        if saidqt:
            tree = cp.parse(sent)
            for subtree in tree.subtrees():
                if subtree.label() == 'QT': 
                    words =  [word[0] for word in subtree]
                    words = " ".join(words)
                    #print words
                    return words
```

```python
#flag all the paragraphs that contain said and then dump them into into main dataframe
bqs['said_sent'] = pd.DataFrame(bqs.text.str.contains('said+'))
```

```python
results = []
said_story_ids = []
for row in df.T.iteritems():
  peep = chunktext(row[1]['sentences'], person_said_rule)
  if peep:
      results.append(peep)
  else:
     results.append('')
  pid = row[1]['story_id'] 
  said_story_ids.append(pid)

```

```python
#now merge the data back into the dataframe.
said_stuff = np.array(zip(said_story_ids, results))
df_said_stuff = pd.DataFrame({'story_id':said_stuff[:,0],'said_speaker':said_stuff[:,1]})
```
